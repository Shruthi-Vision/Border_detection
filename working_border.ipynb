```python
def Border_detection(image_no,input_path,output_path,coord1,coord2):
        #reading the image 
        img = cv2.imread(input_path, 2) 
        #converting to binary
        ret, bw_img = cv2.threshold(img, 40, 170, cv2.THRESH_BINARY) 
       
        bw = cv2.threshold(img, 40, 255, cv2.THRESH_BINARY) 

        # detect the contours on the binary image 
        contours, hierarchy = cv2.findContours(bw_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                                            
        # draw contours on the original image
        image_copy = img.copy()
        
        cv2.drawContours(image_copy, contours, -1, (255, 0, 0), 2)  
        
        #setting the output path
        name = output_path + str(image_no) + '.bmp'
        
        for contour in contours:
            area = cv2.contourArea(contour)
            rect = cv2.minAreaRect(contour)
            box = cv2.boxPoints(rect)
            if area>600000:
                # Find bounding rectangles
                x,y,w,h = cv2.boundingRect(contour)
                # Draw the rectangle
                cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),1)
                print(x,x+w)
                #writing the output
                cv2.imwrite(name, img)

                #checking the distance between the x (can border) and xmin (label value)
                
                (xmin,ymin)=coord1
                (xmax,ymax)=coord2
                distance=(x+w)-xmax  
                print(distance)
                #set up the threshold value based on testing
                if distance >1275 or distance <100:
                    print('Fail')
                else:
                    print('Pass')               
                
                
```
